```python
import seaborn as sns
```

```python
#Basic plotting with Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
#create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot( x = 'total_bill' , y = "tip" , data = tips )
plt.title("Scatter plot")
plt.show()
```

```python
#line plot

sns.lineplot(x='size' , y = 'total_bill' , data = tips)
```

```python
#Categorical Plots
## Bar plot
sns.barplot(x = 'day' , y = 'total_bill' , data =tips)
plt.title('Bar plot of Total Bill by Day')
plt.show()
```

```python
#box plot
sns.boxplot(x='day' , y = 'total_bill' , data=tips)
```

```python
# violin plot

sns.violinplot(x = 'day' , y = 'total_bill' , data=tips)
```

```python
#histogram
sns.histplot(tips['total_bill'] , bins = 10 , kde=True)
```

```python
#KDE Plot

sns.kdeplot(tips['total_bill'] , shade =True)
```

```python
#Pariplot
sns.pairplot(tips)
```

```python
#Heatnmap
corr = tips[['total_bill' , 'tip' , 'size']].corr()
sns.heatmap(corr , annot= True , cmap='coolwarm')
```

```python
#pracetice
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

dates = pd.date_range(start='2023-01-01', periods=100, freq='D')

categories = ['A', 'B', 'C']

df = pd.DataFrame({
    'Date': dates,
    'Category': np.random.choice(categories, size=100),
    'Value': np.random.randint(10, 100, size=100),          
    'Score': np.random.normal(loc=50, scale=10, size=100)    
})


df

```

```python
sns.histplot(df['Score'] , bins = 20 , kde=True  )
```

```python

std = np.std(df['Score'])

sns.kdeplot(df['Score'], shade=True)

plt.title("Score Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Score", fontsize=12)
plt.ylabel("Density", fontsize=12)

# เพิ่มข้อความแสดงค่า standard deviation ด้วยฟอร์แมตที่ดูดี
# ปรับตำแหน่งข้อความและใส่ background box ให้ข้อความเด่นขึ้น
plt.text(50 , 0.02, f"Std: {std:.2f}\n Mean : {np.mean(df['Score']):.2f}", fontsize=12, color='red',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='red'))

plt.show()
```
